```
#!/usr/bin/env python3
from sympy import pprint, symbols, Integral, plot, sin, cos, pi

print('9.')

x = symbols('x')
f = sin(2 * x) ** 2 * cos(2 * x) ** 2
I = Integral(f, (x, 0, pi / 2))
for t in [I, I.doit()]:
    pprint(t)
    print()

p = plot(f, 1 / 2 * sin(x) ** 2 * cos(x) ** 2, legend=True, show=False)
colors = ['red', 'green']
for i, color in enumerate(colors):
    p[i].line_color = color
p.save('sample9.svg')
```
