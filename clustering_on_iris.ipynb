```
import pandas as pd
import seaborn as sns
from sklearn import datasets
df = sns.load_dataset("iris")
print(df.head())
```

```
from sklearn import cluster
iris=datasets.load_iris()
data=iris['data']
model=cluster.KMeans(n_clusters=3)
model.fit(data)
from matplotlib import pyplot as plt
labels=model.labels_
ldata=data[labels==0]
plt.scatter(ldata[:,2],ldata[:,3],c='teal',alpha=0.3,s=100,marker='1')
ldata=data[labels==1]
plt.scatter(ldata[:,2],ldata[:,3],c='darkgreen',alpha=0.3,s=100,marker='2')
ldata=data[labels==2]
plt.scatter(ldata[:,2],ldata[:,3],c='red',alpha=0.3,s=100,marker='x')
```

```
from matplotlib import pyplot as plt
from sklearn import cluster
from sklearn import datasets
iris=datasets.load_iris()
data=iris['data']
model=cluster.KMeans(n_clusters=3)
model.fit(data)
labels=model.labels_
MARKERS=["v","^","+","x","d","p","s","1","2"]
def scatter_by_features(feat_idx1,feat_idx2):
  for lbl in range(labels.max()+1):
    clustered=data[labels==lbl]
    plt.scatter(clustered[:,feat_idx1],clustered[:,feat_idx2],c='crimson',alpha=0.3,s=100,marker=MARKERS[lbl],label='label{}'.format(lbl))
  plt.xlabel(iris['feature_names'][feat_idx1],fontsize='large')
  plt.ylabel(iris['feature_names'][feat_idx2],fontsize='large')
plt.figure(figsize=(16,16))
plt.subplot(3,2,1)
scatter_by_features(0,1)
plt.subplot(3,2,2)
scatter_by_features(0,2)
plt.subplot(3,2,3)
scatter_by_features(0,3)
plt.subplot(3,2,4)
scatter_by_features(1,2)
plt.subplot(3,2,5)
scatter_by_features(1,3)
plt.subplot(3,2,6)
scatter_by_features(2,3)
plt.show()
```
