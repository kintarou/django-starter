```
import pandas as pd
import seaborn as sns
from sklearn import datasets
df = sns.load_dataset("iris")
print(df.head())
```

```
from sklearn import cluster
iris=datasets.load_iris()
data=iris['data']
model=cluster.KMeans(n_clusters=3)
model.fit(data)
from matplotlib import pyplot as plt
labels=model.labels_
ldata=data[labels==0]
plt.scatter(ldata[:,2],ldata[:,3],c='teal',alpha=0.3,s=100,marker='o')
ldata=data[labels==1]
plt.scatter(ldata[:,2],ldata[:,3],c='darkgreen',alpha=0.3,s=100,marker='^')
ldata=data[labels==2]
plt.scatter(ldata[:,2],ldata[:,3],c='red',alpha=0.3,s=100,marker='*')
```
