```
import seaborn as sns
sns.set_style('whitegrid')
tips=sns.load_dataset('tips')
tips
```

```
sns.lmplot(x='total_bill',y='tip',hue='smoker',data=tips)
```

```
sns.lmplot(x='total_bill',y='tip',col='smoker',data=tips)
```

```
sns.lmplot(x='total_bill',y='tip',row='smoker',data=tips)
```

```
sns.lmplot(x='total_bill',y='tip',hue='smoker',data=tips,markers=[",","^"])
```

```
sns.lmplot(x='total_bill',y='tip',hue='sex',data=tips,markers=[",","^"],order=3)
```

```
sns.lmplot(x='total_bill',y='tip',hue='smoker',data=tips,markers=[",","^"],palette=dict(Yes="b",No="r"),order=3)
```

```
sns.lmplot(x='total_bill',y='tip',hue='sex',data=tips,markers=[",","^"],palette=dict(Female="b",Male="r"),order=3)
```
