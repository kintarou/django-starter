```
# coding: utf-8
import numpy as np
import matplotlib.pyplot as plt


def sigmoid(x):
    return 1 / (1 + np.exp(-x))


def ReLU(x):
    return np.maximum(0, x)


def tanh(x):
    return np.tanh(x)
    
x = np.random.randn(1000, 100)  # 1000蛟九・繝・・繧ｿ
node_num = 100  # 蜷・國繧悟ｱ､縺ｮ繝弱・繝会ｼ医ル繝･繝ｼ繝ｭ繝ｳ・峨・謨ｰ
hidden_layer_size = 5  # 髫繧悟ｱ､縺・螻､
activations = {}  # 縺薙％縺ｫ繧｢繧ｯ繝・ぅ繝吶・繧ｷ繝ｧ繝ｳ縺ｮ邨先棡繧呈ｼ邏阪☆繧・


for i in range(hidden_layer_size):
    if i != 0:
        x = activations[i-1]

    w = np.random.randn(node_num, node_num) * 1
    # w = np.random.randn(node_num, node_num) * 0.01
    # w = np.random.randn(node_num, node_num) * np.sqrt(1.0 / node_num)
    # w = np.random.randn(node_num, node_num) * np.sqrt(2.0 / node_num)


    z = np.dot(x, w)
    a = sigmoid(z)


    # 豢ｻ諤ｧ蛹夜未謨ｰ縺ｮ遞ｮ鬘槭ｂ螟峨∴縺ｦ螳滄ｨ薙＠繧医≧・・    z = sigmoid(a)
    # z = ReLU(a)
    # z = tanh(a)

    activations[i] = a

# 繝偵せ繝医げ繝ｩ繝繧呈緒逕ｻ
for i, a in activations.items():
    plt.subplot(1, len(activations), i+1)
    plt.title(str(i+1) + "-layer")
    # plt.xlim(0.1, 1)
    # plt.ylim(0, 7000)
    plt.hist(a.flatten(), 30, range=(0,1))
plt.show()
```
