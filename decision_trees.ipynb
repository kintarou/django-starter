```
import numpy as np
import matplotlib.pyplot as plt

# 表の出る確率
P1 = np.linspace(0, 1, 1000)

# 裏の出る確率
P2 = 1 - P1

# エントロピー
I = lambda p1, p2: -p1 * np.log(p1) - p2 * np.log(p2)

# エントロピーの値の計算
result = np.array([I(p1, p2) for p1, p2 in zip(list(P1), list(P2))])
result[np.isnan(result)] = 0

# 可視化
plt.scatter(P1, result)
plt.show()
```
