```
#k-means
from sklearn import cluster
from matplotlib import pyplot as plt
from sklearn import datasets
diabetes = datasets.load_diabetes()
data = diabetes['data']

model = cluster.KMeans(n_clusters=3)
model.fit(data)

labels = model.labels_

ldata = data[labels == 0]
plt.scatter(ldata[:,2], ldata[:,3], c='black', alpha=0.3, s=100, marker='o')

ldata = data[labels == 1]
plt.scatter(ldata[:,2], ldata[:,3], c='red', alpha=0.3, s=100, marker='^')

ldata = data[labels == 2]
plt.scatter(ldata[:,2], ldata[:,3], c='blue', alpha=0.3, s=100, marker='*')

plt.xlabel(diabetes['feature_names'][2])
plt.ylabel(diabetes['feature_names'][3])

plt.show()
```

```
#dendrogram
from sklearn import cluster
from matplotlib import pyplot as plt
from sklearn import datasets
diabetes = datasets.load_diabetes()
data = diabetes['data']

model = cluster.AgglomerativeClustering(n_clusters=3, linkage='ward')
model.fit(data)

labels = model.labels_

ldata = data[labels == 0]
plt.scatter(ldata[:,2], ldata[:,3], c='black', alpha=0.3, s=100, marker='o')

ldata = data[labels == 1]
plt.scatter(ldata[:,2], ldata[:,3], c='red', alpha=0.3, s=100, marker='^')

ldata = data[labels == 2]
plt.scatter(ldata[:,2], ldata[:,3], c='blue', alpha=0.3, s=100, marker='*')

plt.xlabel(diabetes['feature_names'][2])
plt.ylabel(diabetes['feature_names'][3])

plt.show()
```
